```python
# Dictionaries
# Unlike Sets and lists, dictionaries store pairs of elements: keys and values.
# E.g. Consider a dictionary with keys are element names and the values are their corresponding atomic numbers.
elements = {'hydrogen': 1, 'helium': 2, 'carbon': 6}

# look up values in the dictionary using square brackets enclosing a key:
print(elements['carbon'])

#insert new values into the dictionary with square brackets:
elements['lithium'] = 3
print(elements['lithium'])

print(elements)

# Dictionary keys are similar to list indices: we can select elements from the data structure by putting the index/key in square brackets.
# Dictionaries have keys of any immutable type.

# Define a Dictionary, population,
# that provides information
# on the world's largest cities.
# The key is the name of a city
# (a string), and the associated
# value is its population in
# millions of people.

#   Key     |   Value
# Shanghai  |   17.8
# Istanbul  |   13.3
# Karachi   |   13.0
# Mumbai    |   12.5

population = {'Shanghai':17.8, 'Istanbul':13.3, 'Karachi':13.0,'Mumbai': 12.5}

# What's in the Dictionary?
# 1. using 'in'

if 'Mumbai' in population:
    print("That's a real key!")
else:
    print("There's no such key!")
    
# 2. Square Brackets     

print(population['Mumbai'])
# print(population['California'])  # if key not in dictionary, returns Error; 'Keyerror'

# 3. Using 'get'

print(population.get('Mumbai'))
print(population.get('California')) # if key not in dictionary, returns 'None'
elements.get('California', 'There\'s no such element!') # if key not in dictionary, returns 'String'
```
